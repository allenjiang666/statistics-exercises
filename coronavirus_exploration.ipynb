```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
coronavirus = pd.read_csv('NUMBERS_data.csv')

coronavirus = coronavirus.rename(columns={"Countries and territories": "Country"}).drop(columns= "Total Conf Cases all")
```

```python
coronavirus_in_china =  coronavirus[coronavirus.Country == "China"]
coronavirus_in_china = coronavirus_in_china.reset_index(drop=True)
coronavirus_in_china
```

```python
coronavirus_in_china['cumulative_cases'] = coronavirus_in_china.Cases.cumsum()

coronavirus_in_china[['Date', 'cumulative_cases']]
```

```python
plt.figure(figsize = (16, 9))
x = coronavirus_in_china.Date
y = coronavirus_in_china.cumulative_cases

plt.plot(x, y)
```
